## Grouping with Occupation Dataset

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
```

```python
users = pd.read_table(url, sep='|', index_col='user_id')
```

```python
users.head()
```

### Discover what is the mean age per occupation

```python
users.groupby('occupation')['age'].agg(['mean'])
```

###  Discover the Male ratio per occupation and sort it from the most to the least

```python
ans = users[users['gender'] == 'M'].groupby('occupation')['gender'].count() / users['occupation'].value_counts() * 100

ans.sort_values(ascending=False)
```

### For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation')['age'].agg(['min', 'max'])
```

### For each occupation present the percentage of women and men

```python
occ_counts = users['occupation'].value_counts()

pct_women = round(users[users['gender'] == 'F'].groupby(['occupation'])['gender'].count() / occ_counts * 100, 2)
pct_men = round(users[users['gender'] == 'M'].groupby(['occupation'])['gender'].count() / occ_counts * 100, 2)
```

```python
pd.concat([pct_women, pct_men], axis=1).rename(columns={0: 'pct_women', 1: 'pct_men'})
```
