```python
import argparse
import requests
import math

class NYTimesSource(object):
    def __init__(self):
        pass
    
    def getDataBatch(self, batch_size):
        """
        Generator - Get data from source on batches.

        :returns One list for each batch. Each of those is a list of
                 dictionaries with the defined rows.
        """
        # TODO: implement - this dummy implementation returns one batch of data
        URL = 'https://api.nytimes.com/svc/search/v2/articlesearch.json'

        full_url = URL + '?' + 'api-key=' + self.args.api_key + '&q=' + self.args.query + '&page='
        pages_to_query = int(math.ceil(batch_size / 10))  #10 is the max results per page
        result = []
        page = 0
        max_pages = 1000
        hit_pages = math.ceil(requests.get(full_url + str(0)).json()['response']['meta']['hits'] / 10)
        print(max_pages, hit_pages)
        
        while page < min(max_pages, hit_pages) :
            for _ in range(pages_to_query):
                response = requests.get(full_url + str(page)).json()
                result.extend(response['response']['docs'])
                page += 1
                
            print(len(result))
            batch_result = [result.pop() for idx in range(batch_size)]
            yield batch_result
    
        
```

```python
def flatten2(dict_obj, sep='.'):
    flattened = dict()
    
    def traverse_keys(level, flattened_key=''):
        if isinstance(level, dict):
            for key, val in level.items():
                traverse_keys(val, flattened_key + sep + key if flattened_key != '' else key)
        elif isinstance(level, list):
            for i in range(len(level)):
                traverse_keys(level[i], flattened_key + sep + str(i) if flattened_key != '' else str(i))
        else:
            flattened[flattened_key] = level
            
    traverse_keys(dict_obj)
    
    return flattened
```
